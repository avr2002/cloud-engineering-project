```python
import requests
import time

# localhost URL
# url = "http://localhost:8000/v1/files"
# N = 10

# API Gateway endpoint URL
url = "https://j0o3bozns6.execute-api.ap-south-1.amazonaws.com/prod/v1/files"

# MLOps TA URL
# url = "https://3rbeuvjyd3.execute-api.us-west-2.amazonaws.com/prod/v1/files"
# N = 1000000

TEST_FILE_CONTENT = b"Dummy file content for testing."
TEST_FILE_CONTENT_TYPE = "text/plain"


def generate_file_path(i) -> str:
    return f"file-path-{i}.txt"

def put_request(file_path):
    files = {
        "file_content": (file_path, TEST_FILE_CONTENT, TEST_FILE_CONTENT_TYPE)
    }
    response = requests.put(f"{url}/{file_path}", files=files)
    if response.status_code not in [200, 201]:
        print(f"PUT Status Code: {response.status_code}, Response: {response.headers}")

def get_request(file_path):
    response = requests.get(f"{url}/{file_path}")
    if response.status_code != 200:
        print(f"GET Status Code: {response.status_code}, Response: {response.headers}")

def delete_request(file_path):
    response = requests.delete(f"{url}/{file_path}")
    if response.status_code != 204:
        print(f"DELETE Status Code: {response.status_code}, Response: {response.headers}")

# Loop to call endpoints 300 times
# for i in range(1, N+1):
#     file_path = generate_file_path(i)    # "file-path.txt" # 
#     # if i % 10 == 0:
#     #     print(f"Iteration: {i}, File Path: {file_path}")
#     put_request(file_path)
#     get_request(file_path)
#     delete_request(file_path)
#     # Optional: Add a delay between requests to avoid overwhelming the server
#     time.sleep(1)  # Delay in seconds

# print(f"Completed {N} iterations of API calls.")

i = 0
while True:
    i += 1
    file_path = "file-path.txt"
    put_request(file_path)
    get_request(file_path)
    delete_request(file_path)
    if i % 10 == 0:
        print(f"Iteration: {i}")
    
    time.sleep(1)  # Delay in seconds
```
